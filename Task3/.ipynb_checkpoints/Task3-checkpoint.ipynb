```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, recall_score, roc_auc_score
```

```python
features = ['credit_lines_outstanding', 'debt_to_income', 'payment_to_income', 'years_employed', 'fico_score']
loans_df['payment_to_income'] = loans_df['loan_amt_outstanding'] / loans_df['income']
loans_df['debt_to_income'] = loans_df['total_debt_outstanding'] / loans_df['income']
```

```python
model = LogisticRegression(random_state = 1024, solver='liblinear', tol=1e-5, max_iter = 10000).fit(loans_df[features], loans_df['default'])
print(model.coef_, model.intercept_)
y_pred = model.predict(loans_df[features])
fpr, tpr, thresolds = metrics.roc_curve(loans_df['default'], y_pred)
print((1.0 * (abs(loans_df['default'] - y_pred)).sum()) / len(loans_df))
print(metrics.auc(fpr,tpr))
```
